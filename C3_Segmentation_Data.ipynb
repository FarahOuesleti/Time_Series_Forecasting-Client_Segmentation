Function to Determine **non NAN** values in a column

```python
def NonNanValues(DF,columnName):
  L=[]
  for i in range(152) :
    if not (DF[columnName].isna()[i]) :
      L.append(i)
  #print(f' liste : {L}')
  return L

# output : 32 , 107
```
